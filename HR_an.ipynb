```python
ohe_columns = ['dept', 'last_year_promo', 'last_year_violations']
ord_columns = ['level', 'workload']
num_columns = ['employment_years', 'supervisor_evaluation', 'salary']

ohe_pipe = Pipeline(
    [
        (
            'simple_imputer_ohe_1',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
                (
            'simple_imputer_ohe_2',
            SimpleImputer(missing_values=' ', strategy='most_frequent')
        ),
        (
            'ohe',
            OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
)

ord_pipe = Pipeline([
    ('ordinal', OrdinalEncoder(
        handle_unknown='use_encoded_value',
        unknown_value=-1,
        categories=[
            ['junior', 'middle', 'senior'],
            ['medium', 'high', 'low']
        ]
    )),
    ('simpleImputer_after_ord', SimpleImputer(strategy='most_frequent'))
])
num_pipe = Pipeline([
    ('simple_imputer_num', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

data_preprocessor = ColumnTransformer(
    transformers=[
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', num_pipe, num_columns)
    ],
    remainder='passthrough'
)
```

## Шаг 5. Обучение моделей

Будем искать модель с лучшими параметрами. В задании скзано, что модель нужно будет оценить новой метрикой SMAPE. Для ее расчета попросили сделать функцию. В теории про эту метрику ничего не было, но почитав инструкции стало понятно, что метрика довольно полезная. SMAPE измеряет среднее абсолютное процентное отклонение между прогнозами и фактическими значениями.

```python
def smape(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred) / ((np.abs(y_true) + np.abs(y_pred)) / 2)) * 100

smape_scorer = make_scorer(smape, greater_is_better=False)
```

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeRegressor(random_state=RANDOM_STATE))
])
```

```python
ohe_columns = ['dept', 'last_year_promo', 'last_year_violations']
ord_columns = ['level', 'workload']
num_columns = ['employment_years', 'supervisor_evaluation', 'salary', 'job_satisfaction_rate_predicted']

ohe_pipe = Pipeline([
    ('simpleImputer_ohe', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])

ord_pipe = Pipeline([
    ('ordinal', OrdinalEncoder(
        handle_unknown='use_encoded_value',
        unknown_value=-1,
        categories=[
            ['junior', 'middle', 'senior'],
            ['medium', 'high', 'low']
        ]
    )),
    ('simpleImputer_after_ord', SimpleImputer(strategy='most_frequent'))
])

data_preprocessor = ColumnTransformer(
    transformers=[
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ],
    remainder='passthrough'
)
```
